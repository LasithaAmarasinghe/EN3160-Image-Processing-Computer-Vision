# Create Image

```python
import cv2 as cv
import numpy as np
import matplotlib.pyplot as plt

im =np.zeros((6,8),dtype=np.uint8)   ## creates a new array of shape (6, 8) filled with zeros.
im[2,3]=255; im[3,5]=255

fig,ax=plt.subplots(1,1,figsize=(6,8))  ## (1, 1) specifies a grid of 1 row and 1 column of subplots.  ## figure (width, height) in inches
ax.imshow(im,cmap='gray',vmin=0,vmax=255)
ax.xaxis.set_ticks_position('top')   ## Moves the x-axis ticks to the top of the plot area
plt.show()
```

```python
import cv2 as cv
import numpy as np
import matplotlib.pyplot as plt

im=np.zeros((6,8,3),dtype=np.uint8)   
# 6: Height of the image (number of rows)  # 8: Width of the image (number of columns) # 3: Depth of the image, represents RGB channels (Red, Green, Blue).

im[2,3]=(255,190,203)
im[4,5]=(101,154,136)

fig,ax=plt.subplots(1,1,figsize=(6,8))
ax.imshow(im)
ax.xaxis.set_ticks_position('bottom')
plt.show()
```
