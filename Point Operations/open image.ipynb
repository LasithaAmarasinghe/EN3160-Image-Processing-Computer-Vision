# Display Image

```python
## image display using matplotlib

import cv2 as cv
import numpy as np
import matplotlib.pyplot as plt

im= cv.imread('img/emma.jpg')  ## attempts to read the image file using opencv
assert im is not None ## checks whether image is loaded successfuly   
## if im is None, AssertionError occurs. it indicates that the image could not be loaded.

fig,ax = plt.subplots(1,2,figsize=(12,8))
ax[0].imshow(cv.cvtColor(im,cv.COLOR_BGR2RGB))  ## Converts the image from BGR format used by OpenCV to RGB format expected by Matplotlib.
ax[1].imshow(im)  ## this doesn't give correct colors since BGR is displayed as RGB.
ax[0].set_title('Correct')
ax[1].set_title('Wrong')
ax[0].axis('off')
ax[1].axis('off')
plt.show()
```

```python
## image display using opencv

import cv2 as cv

im= cv.imread('img/emma.jpg')
cv.namedWindow('Image',cv.WINDOW_AUTOSIZE)   ## cv.namedWindow(window_name, flags)  
## cv.WINDOW_AUTOSIZE flag -> OpenCV sets the window size automatically to match the dimensions of the image displayed in the window
cv.imshow('Image',im)
cv.waitKey(0)
cv.destroyAllWindows() ## clean up and close all OpenCV opened windows
```

# Image Properties

```python
## display image properties

import cv2 as cv
import matplotlib.pyplot as plt

im=cv.imread('img/emma.jpg')
fig,ax=plt.subplots(figsize=(6,8))
ax.imshow(cv.cvtColor(im,cv.COLOR_BGR2RGB))
ax.xaxis.set_ticks_position('top')
plt.show()

print('Image Shape:',im.shape)     ## dimensions (height, width, channels)     ## channels=3 for RGB images
print('Image Data Type:',im.dtype) ## data type of the elements in image array   ## uint8 -> unsigned 8-bit integers
print('Image Size:',im.size)       ## total number of elements (pixels)   ## for color images, im.size = height * width * channels
```
